```python
amount_due=50

while amount_due>0:
    print("Amount Due=",amount_due)
    coin=input("Enter coin")

    if coin=="25" or coin=="10" or coin=="5":
        amount_due=amount_due-int(coin)   
    else:
        print("Invalid coin")

print("Change Owed=",abs(amount_due))
```

```python
text=input("Enter a string: ")
word=""
for i in text:
    if ord(i)>=65 and ord(i)<=90:
        word+="_"+i.lower()
    else:
        word+=i
print(word)
```

```python
text=input("Enter a string: ")
word=""

for ch in text:
    if ch == "_":
        replace=ch.replace("_","")
        word+=replace.upper()

    else:        
        word+=ch
print(word)    
```

```python
text = input("Enter text: ")

result = ""
uppercase_next = False

for ch in text:
    if ch == "_":
        uppercase_next = True
        continue
    if uppercase_next:
        result += ch.upper()
        uppercase_next = False
    else:
        result += ch

print(result)
```

```python
text = input("Enter text: ")
word=" "
uppercase_next=False

for ch in text:
    if ch=="-":
        uppercase_next=True
    elif uppercase_next:
        word+=ch.upper()
        uppercase_next=False
    else:
        word+=ch

print(word)
```

```python
class flight():
    def __init__(self,capacity):
        self.capacity=capacity
        self.passenger=[]
    
    def add_passenger(self, name):
        if not self.open_seats():
            return False

        self.passenger.append(name)
        return True
    def open_seats(self):
        self.capacity-len(self.passenger)

flight=Flight(3)

people=["Alice","Bob","Moty","Richard"]
for person in people:
    success=flight.add_passenger(person)
    if success:
        print(f"Added {person} to flight Successfully")
    else:
        print(f"{person} Not Added to Flight")
```

```python
import sys
try:
    x=int(input("x: "))
    y=int(input("y: "))
except ValueError:
    print("Error: Invalid input")
    sys.exit(1)
try:
    result = x / y
except ZeroDivisionError:
    print("Cannot divide by 0")
    sys.exit(1) 

print(f"{x}/{y}={result}")
```
