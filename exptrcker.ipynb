```python
import os
import datetime

def add_expense():
    while True:
        amount = input("Enter amount => ")
        try:
            float(amount)
            break
        except ValueError:
            print("Oops! Please enter a valid number (e.g., 150 or 50.5).")

    category = input("Enter Category => ")
    date = datetime.date.today().strftime("%Y-%m-%d")
    with open("expenses2.txt", "a") as file:
        file.write(f"{date},{amount},{category}\n")
    print("Expenses added successfully")


def view_expense():
    try:
        with open("expenses2.txt", "r") as file:
            data = file.readlines()

        print("Your Expenses:")
        for line in data:
            date, amount, category = line.strip().split(",")
            print(f"{date}: Rs {amount} => {category}")
    except FileNotFoundError:
        print("No expense found")


def total_expenses():
    try:
        with open("expenses2.txt", "r") as file:
            data = file.readlines()
            total = 0
            for line in data:
                _, amount, _ = line.strip().split(",")
                total += float(amount)
            print("Total Expenses: RS", total)
    except FileNotFoundError:
        print("No expense found")


def clear_expense():
    choice = input("Are you sure you want to clear all expenses? (yes/no) => ")
    if choice.lower() == "yes":
        if os.path.exists("expenses2.txt"):
            os.remove("expenses2.txt")
            print("File deleted")
        else:
            print("No expense found")


while True:
    print("======= Smart Expense Tracker =======")
    print("1. Add Expense")
    print("2. View Expense")
    print("3. Total Expenses")
    print("4. Clear Expenses")
    print("5. Exit")
    choice = input("Enter your choice (1-5) => ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expense()
    elif choice == "3":
        total_expenses()
    elif choice == "4":
        clear_expense()
    elif choice == "5":
        print("Exiting... Goodbye!")
        break
    else:
        print("Invalid choice")
```
