```python
import os
import datetime

FILE_NAME="expenses2.txt"
def add_expense():
    while True:
        try:
            amount = float(input("Enter amount => "))
            break
        except ValueError:
            print("Oops! Please enter a valid number (e.g., 150 or 50.5).")

    category = input("Enter Category => ")
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(FILE_NAME, "a") as file:
        file.write(f"{timestamp},{amount},{category}\n")
    print("Expenses added successfully")


def view_expense():
    try:
        with open(FILE_NAME, "r") as file:
            data = file.readlines()
        if not data:
            print("No expense recorded yet.")
            return
        print("Your Expenses:")

        for line in data:
            if not line.strip():
                continue
            timestamp, amount, category = line.strip().split(",")
            print(f"{timestamp}: Rs {amount} => {category}")
    except FileNotFoundError:
        print("No expense found")


def total_expenses():
    try:
        with open(FILE_NAME, "r") as file:
            data = file.readlines()
            total = 0.0

            for line in data:
                if not line.strip():
                    continue
                _, amount, _ = line.strip().split(",")
                total += float(amount)
            print("Total Expenses: RS", total)
    except FileNotFoundError:
        print("No expense found")

def view_category_expenses():
    try:
        with open(FILE_NAME, "r") as file:
            data=file.readlines()
            category_expenses={}
            for line in data:
                if not line.strip():
                    continue
                

def clear_expense():
    choice = input("Are you sure you want to clear all expenses? (yes/no) => ")
    if choice.lower() == "yes":
        if os.path.exists(FILE_NAME):
            os.remove(FILE_NAME)
            print("File deleted")
        else:
            print("No expense found")


while True:
    print("======= Smart Expense Tracker =======")
    print("1. Add Expense")
    print("2. View Expense")
    print("3. Total Expenses")
    print("4. Clear Expenses")
    print("5. Exit")
    choice = input("Enter your choice (1-5) => ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expense()
    elif choice == "3":
        total_expenses()
    elif choice == "4":
        clear_expense()
    elif choice == "5":
        print("Exiting... Goodbye!")
        break
    else:
        print("Invalid choice")
```
