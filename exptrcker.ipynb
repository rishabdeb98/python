```python
#smart exp tracker

def add_expense():
    amount=input("Enter amount = > ")
    category=input("Enter Category (food /travel/shopping)=> ")

    with open("expenses.txt", 'a') as file:
        file.write(amount + "," + category+"\n")

    print("Expense Added successfully")

def view_expense():
    try:
        with open("expenses.txt", 'r') as file:
            data=file.readlines()

        print("\nYour Expenses")
        for line in data:
            amount,category=line.strip().split(",")
            print("RS",amount,"=>",category)
    except FileNotFoundError:
        print("No expense Found")
def tatal_expenses():
    try:
        with open("expenses.txt", 'r') as file:
            data=file.readlines()

        total=0 
        for line in data:
            amount,category=line.strip().split(",")
            total+=float(amount)
        print("\nTotal Expenses: RS",total,"\n")
    except FileNotFoundError:
        print("No expense Found")

while True:
    print("======= Smart Expense Tracker =======")
    print("1. Add Expense")
    print("2. view Expense")
    print("3. Total Expenses")
    print("4. Exit")
    choice=input("Enter your choice (1-4)=>")
    if choice=="1":
        add_expense()
    elif choice=="2":
        view_expense()
    elif choice=="3":
        tatal_expenses()
    elif choice=="4":
        print("Exiting... Goodbye!")
        break
    else:
        print("Invalid choice")


            
```
